# Illustrating Transfer Learning on VGG

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
from keras.applications.vgg16 import VGG16
from keras.models import Model
from keras.layers import Dense, Flatten

print("Tensorflow Version: {}".format(tf.version.VERSION))
```

```python
# load a pretrained model
model = VGG16()
model.summary()
```

# The VGG feature-map extractor
to output the last (convolutional) activity map

```python
# load model without output layerS (model-dependent)
model = VGG16(include_top=False)
model.summary()
```

# The VGG feature-vector extractor

```python
# load model without output layers and add a global pooling average
model = VGG16(include_top=False, pooling='avg')
model.summary()
```

# VGG for a different number of classes (not pretrained)

```python
# load model and change output classes (implies discarding weights)
model = VGG16(weights=None,classes=10)
model.summary()
```

# VGG without the last dense layer softmax



```python
model = VGG16()
# remove JUST the LAST output layer 
model = Model(inputs=model.inputs, outputs=model.layers[-2].output)
model.summary()
```

# Pretrained VGG for a different set of classes

```python
# load model without classifier layers
model = VGG16(include_top=False, input_shape=(300, 300, 3))

# add new classifier layers
flat1 = Flatten()(model.layers[-1].output)
class1 = Dense(1024, activation='relu')(flat1)
output = Dense(10, activation='softmax')(class1)
# define new model
model = Model(inputs=model.inputs, outputs=output)
model.summary()
```

```python
model.get_layer('block1_conv1').trainable
```

# Prepare pretrained VGG for a different set of classes and partial retraining

```python
# load model without classifier layers
model = VGG16(include_top=False, input_shape=(300, 300, 3))

for l in model.layers:
  l.trainable = False

# add new classifier layers
flat1 = Flatten()(model.layers[-1].output)
class1 = Dense(1024, activation='relu')(flat1)
output = Dense(10, activation='softmax')(class1)
# define new model
model = Model(inputs=model.inputs, outputs=output)
model.summary()
```

```python
for l in model.layers:
  print('%20s, %10s'%(l.name,l.trainable))
```
