# Some Famous and Known Convolutional Models

```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model

print("Tensorflow Version: {}".format(tf.version.VERSION))
```

#VGG
developed by the Visual Graphics Group at Oxford in 2014.

VGG-16 and VGG-19 models are available.

```python
from keras.applications.vgg16 import VGG16

model = VGG16()
model.summary()
```

```python
x = model.get_layer('predictions')
print(x.activation)

x = model.get_layer('block5_conv2')
print(x.activation)
```

```python
plot_model(model, to_file='model.png', 
           show_shapes=True, show_layer_activations=True,
           rankdir='LR', #TB by default
           dpi=54)
```

# GoogLeNet
